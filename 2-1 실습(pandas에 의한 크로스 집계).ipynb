파이썬 코드에서 u 접두어는 유니코드 문자열(Unicode string)을 나타냅니다.<br>그러나 Python 3 버전에서는 u 접두어를 사용하지 않고도 문자열이 기본적으로 유니코드로 처리됩니다.

```python
import pandas as pd
df1=pd.read_excel('superstore.xlsx','고객정보')
df1.head()
```

```python
df2=pd.read_excel('superstore.xlsx','주문')
df2.head()
```

```python
df3=pd.merge(df1,df2,on='고유 ID')
df3.head()
```

```python
#지역별, 범주별 수량 크로스 집계
df3.pivot_table('수량',['지역','시도'],'범주', aggfunc='sum')
```

1. row['지역']: 각 행에 대한 '지역' 열의 값을 가져옵니다.
2. {'수도권': '수도권'}.get(row['지역'], '그 외'): 딕셔너리를 사용하여 '수도권'이면 '수도권'을, 그렇지 않으면 '그 외'를 반환합니다.
3. 이 함수를 df1.apply(category, axis=1)로 사용하여 데이터프레임의 각 행에 적용합니다.
4. 결과로 '지역' 열이 '수도권' 또는 '그 외'로 카테고리화된 새로운 '지역' 열이 생성됩니다.

```python
#지역을 수도권 / 그외로 카테고리화
#지역 카테고리를 함수로 설정기
def category(row):
    return {'수도권':'수도권'}.get(row['지역'],'그 외')
df1['지역']=df1.apply(category, axis=1)
df1
```

```python
from pandas import DataFrame
vtable = {'Uid':[101,101,101,102,102,102], 'Key':['c1','c2','c3','c1','c2','c3'], 'Value':[11,12,13,21,22,23]}
vtable=DataFrame(vtable)
vtable
```

```python
htable = {'Uid':[101,102], 'c1':[11,12], 'c2':[12,22],'c3':[13,23]}
htable=DataFrame(htable)
htable
```

```python
from pandasql import sqldf
query = "SELECT uid, key, sum(value) value FROM vtable GROUP BY 1,2"
df_vtable = sqldf(query)
df_vtable
```

```python
# 피벗
df_vtable.pivot('Uid','Key','value')
```

```python
#피벗
df_vtable.pivot(index='Uid', columns='Key', values='value')
```

```python
#언피벗
htable.melt('Uid', var_name='Key', value_name='value')
```
